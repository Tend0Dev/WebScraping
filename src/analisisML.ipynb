## Importaciones

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_excel('productos_mercadolibre.xlsx')
```

```python
data['Precio'] = data['Precio'].replace({',': ''}, regex=True).astype(float)
```

## 1. Cuales son las marcas mas vendidas

```python
brand_counts = data['Marca'].value_counts()

# Gráfico de las 10 marcas más vendidas
plt.figure(figsize=(10, 6))
brand_counts.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Marcas Más Vendidas')
plt.xlabel('Marca')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45)
plt.show()
```

## 2. Que marca es la mas cara

```python
# Calculamos el precio promedio por marca
average_price_per_brand = data.groupby('Marca')['Precio'].mean().sort_values(ascending=False)
most_expensive_brands = average_price_per_brand.head(5)

# Gráfico de las 5 marcas más caras
plt.figure(figsize=(10, 6))
most_expensive_brands.plot(kind='bar', color='orange')
plt.title('Top 5 Marcas Más Caras')
plt.xlabel('Marca')
plt.ylabel('Precio Promedio')
plt.xticks(rotation=45)
plt.show()
```

## 3. Que marca es la mas barata 

```python
# Recalcular el precio promedio por marca, pero asegúrate de que esté en orden ascendente
average_price_per_brand = data.groupby('Marca')['Precio'].mean().sort_values()
cheapest_brands = average_price_per_brand.head(5)  # Tomar las 5 marcas con precio promedio más bajo

# Gráfico de las 5 marcas más baratas
plt.figure(figsize=(10, 6))
cheapest_brands.plot(kind='bar', color='green')
plt.title('Top 5 Marcas Más Baratas')
plt.xlabel('Marca')
plt.ylabel('Precio Promedio')
plt.xticks(rotation=45)
plt.show()
```

## 4. Rango de promedio del precio sobre carros

```python
# Seleccionamos los 5 carros con el precio más alto
top_5_expensive_cars = data[['Marca', 'Modelo', 'Precio']].sort_values(by='Precio', ascending=False).head(5)

# Gráfico de los 5 carros más caros
plt.figure(figsize=(10, 6))
plt.bar(top_5_expensive_cars['Marca'] + ' ' + top_5_expensive_cars['Modelo'], top_5_expensive_cars['Precio'], color='purple')
plt.title('Top 5 Carros Más Caros')
plt.xlabel('Carro (Marca y Modelo)')
plt.ylabel('Precio')
plt.xticks(rotation=45)
plt.show()
```

## 5. Cuales son las marcas menos vendidas

```python
least_sold_brands = brand_counts[brand_counts == 1]

# Gráfico de las marcas menos vendidas (solo aquellas que tienen 1 venta)
plt.figure(figsize=(10, 6))
least_sold_brands.plot(kind='bar', color='salmon')
plt.title('Marcas Menos Vendidas (Una Venta)')
plt.xlabel('Marca')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=90)
plt.show()
```
